### Tipos de dados e variáveis em Python

Ao programar é importante entender o tipo de dados com que estamos a lidar.

As variáveis podem conter diferentes tipos de dados, e diferentes tipos de dados podem sofrer diferentes operações.

Os tipos de dados nativos do Python são:
- números: inteiros (int), decimais (float) e números complexos (complex)
- texto (str)
- listas (list)
- tuples (tuple)
- conjuntos (set)
- dicionários (dict)

Cada tipo de dado é manipulado usando:
- operadores
- funções
- métodos do tipo de dados (funções *built-in*)

#### Tipo de dados de uma variável

**type()** - aplicando esta função a uma variável, conseguimos obter o tipo de dados que a variável armazena.

```python
x = 5
print(type(x))
```

Ou seja, a variável x é da classe dos números inteiros.

#### Casting: Definição do tipo de dados de uma variável

Em Python não é necessário indicar o tipo de dados de uma variável quando a incializamos, uma vez que herda automaticamente o tipo do do valor que armazena. Contudo, poderão haver momentos em que seja necessário especificar o tipo da variável.

Para tal, é necessário fazer *cast* da variável.

Como Python é uma linguagem orientada a objetos, para definir tipos de dados, as classes são utilizadas através de funções de construtor:

**- int ()** - construir um número inteiro a partir de um número inteiro, um float (arredondamento para baixo) ou um número em modo de texto

**- float ()** - construir um número float a partir de um número inteiro, um número float ou um número em modo de texto

**- complex ()** - construir um número complexo a partir de um número inteiro, um número float ou um número em modo de texto

**- str ()** - construir uma cadeia de texto a partir de um texto ou números

##### Exemplo: **int**()

```python
x = int(1) # x será 1
y = int(2.8) # y será 2
z = int("3") # z será 3
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
```

##### Exemplo: **float()**

```python
x = float(1)     # x será 1.0
y = float(2.8)   # y será 2.8
z = float("3")   # z será 3.0
w = float("4.2") # w será 4.2
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
print(w,type(w))
```

##### Exemplo: complex()

```python
x = complex(1)     # x será 1+0j
y = complex(2.8)   # y será 2.8+0j
z = complex("3")   # z será 3+0j
w = complex("4.2") # w será 4.2+0j
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
print(w,type(w))
```

##### Exemplo: **str()**

```python
x = str("s1") # x será 's1'
y = str(2)    # y será '2'
z = str(3.0)  # z será '3.0'
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
```

#### Variáveis numéricas

##### Tipos de dados numéricos

Há 3 tipos de dados numéricos em Python:

**- inteiros (int):** são inteiros, positivos ou negativos, sem decimais e sem um comprimento fixo

**- 7 dígitos(32-bit) (float):** número positivo ou negativo que contém decimais

**- complexos (complexos):** número complexo, que são escritos com "j" como a parte imaginária

Geramos uma variável numérica quando atribuímos um número a ela

```python
x = 1    #inteiro
y = 2.8  #float
z = 1j   #complexo
```

##### Verificar os tipos de dados

```python
print(type(x))
print(type(y))
print(type(z))
```

##### Conversão entre tipos numéricos

Podemos transformar uma variável numérica entre os 3 tipos de dados usando os métodos **int()**, **float()** ou **complex()**.

- **o método int()** converte uma variável float em int
- **o método float()** converte uma variável int para float
- **o método complex()** converte uma variável int ou float em complexa

**Exceção!** Uma variável complexa não pode ser lançada para int ou float.

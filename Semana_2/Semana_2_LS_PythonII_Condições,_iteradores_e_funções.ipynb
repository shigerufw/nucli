### IF( )

Uma condição envolve o uso do comando **"if"**, que é interpretado como uma **condição "if"** e opera da seguinte forma:
- Quando a condição a ser testada (colocada após o "if") é avaliada como verdadeira (**True**), o código recuado que segue a condição é executado;
- No caso de a condição não ser satisfeita (**False**), o código dentro do bloco da condição não é executado.

Vamos verificar se um número é igual a 10.

```python
a = 5
if (a == 10):
  print("Igual")
else:
  print("Diferente")
```

Neste caso, como **a = 5**, a primeira condição não se verifica, uma vez que **'a' não é igual a 10**. É por isso que é impressa a string **"Diferente"**.

### ELIF( ) = ELSE IF

A função ***elif()*** baseia-se numa condição, que só é executada caso a condição acima ***if()*** não seja cumprida.

```python
a = 3
b = 1.5

if b==a:
    print("b é igual a")
elif b > a:
    print("b é maior que a")
elif b < a:
    print("a é maior que b")
```

Neste caso, **b < a**, logo apenas a terceira condição é verdadeira (sendo aquela que é executada).

O programa testa primeiro a condição do **if**. Uma vez que **b==a** não é verdadeiro, passa para o primeiro **elif**. Visto que, **b > a** também não é verdade, o código avança para o segundo **elif**, sendo este verdadeiro.

### ELSE:

O comando ELSE captura tudo o que não cumpriu as condições previamente testadas.

```python
a = 1
b = 2
if b > a:
    print("b é maior que a")
elif a == b:
    print("a e b sao iguais")
elif b >= a:
    print("b é maior ou igual a a")
else:
    print("a é maior que b")
```

O comando **else** passar por ser uma condição **catch-all**, que é cumprida quando o **if()** ou ***elif()*** anteriores não são satisfeitos.

### Condições aninhadas (Nested conditions)

As condições podem ser aninhadas, ou seja, tendo diversos **ifs** dentro de outros **ifs**.  

**Nota:** Ao implementar condições aninhadas, o cuidado com a identação deve ser redobrado!

```python
x = 25

if x > 10:
    print("sim, " + str(x) + " é maior que 10")

    if x > 20:
        print("sim, " + str(x) + " também é maior que 20!")

        if x > 30:
            print ("sim, " + str(x) + " também é maior que 30")
else:
    print("sim, " + str(x) + " não é maior que 10")
```

## **Loops**

Loops permitem iterar sob código específico, repetindo determinadas tarefas.

Python tem dois comandos para criar loops:
- loop **while** - repete *enquanto* uma certa condição for verdadeira
- loop **for** - repete as tarefas *para cada* elemento de um determinado objeto

**Nota:** Também com os loops é preciso ter cuidado com a identação.

**Nota:** Os loops devem ter sempre uma condição de paragem, ou o código entrará num *loop infinito*.

### Loops while()

Com o loop *while*, o conjunto de tarefas **é executado enquanto** uma determinada condição for verdadeira.

```python
i = 0
while i < 10 : #enquanto o valor de i for inferior a 10, o código abaixo é executado
    print(i) #print i
    i+=1 #aincrementar i
```

Este loop será executado 10 vezes, imprimindo o valor de **i** a cada iteração.Vemos que o loop

Quando **i == 10**, a condição do loop já não é satisfeita e portanto não entrará mais nesse bloco de código e a execução do loop terminará.

### Comando *break*

Com o comando *break*, podemos parar um loop mesmo que a condição associada seja satisfeita

# Funções em Python

Uma função é um **conjunto de instruções** que somente é ativado quando é invocado.

* Essas funções utilizam informações ou variáveis como **entrada**, que são referidas como **parâmetros** da função.

* Para além disso, as funções produzem resultados como **saída** de dados.

### Como criar uma função

Para criar uma função em Python é necessário usar o comando **def**, seguido do **nome** que a dar à função, e com o qual a invocaremos no futuro.

O nome é seguido por parêntesis, dentro dos quais declaramos os inputs ou parâmetros da função.

Os parâmetros representam informações de fora que queremos usar dentro da função.

```python
def minha_funcao():
    print("Imprimir esta mensagem quando a função for invocada")
```

```python
minha_funcao()
```

No exemplo anterior, apenas definimos uma função denominada de *"minha_função"*. Esta função não recebe nenhum parâmetro e apenas imprime uma mensagem.

Após a sua declaração, basta-nos invocá-la através do comando *minha_funcao()*.

### Como invocar uma função

Para chamar uma função, basta escrever o nome da função na linha de comando, com os parêntesiss incluídos (como fizemos no exemplo anterior).

No exemplo seguinte, definimos uma função que recebe um nome como parâmetro e imprime uma mensagem incluindo o esse nome.

```python
def minha_função(nome):
    print("Meu nome é " + nome)
```

Vamos agora invocar a nossa função:

```python
minha_função("sara")
```

### Parâmetros de uma função

Os argumentos de uma função são os dados que fornecemos como entrada à função e que serão processados pelo código contido na função definida.

Esses argumentos são indicados após o nome da função, dentro dos parêntesis. É possível adicionar tantos argumentos quanto forem necessários, sempre separados por vírgulas.

```python
def students(name):
    print("Student name: " + name)
```

#### Parâmetros padrão

Ao definir uma função, podemos determinar um parâmetro padrão. Se ao chamar a função, não passarmos esse argumento, a função funcionará na mesma, pois já tem esse valor padrão pré-definido.

```python
def country(c = "Spain"):
    print("I'm from " + c)

country()
country("France")
country("India")
```

Por outras palavras, se estamos a usar uma lista como argumento para uma função, ela continuará sendo uma lista quando a função for executada.

```python
fruits = ["apple", "banana", "orange"] #global

def my_function(fruits):
  print(fruits)

my_function(fruits)
```

### Recursividade

Python também suporta recursividade, o que implica que uma função pode chamar a si mesma.

Essa abordagem permite-nos percorrer os dados até atingirmos o resultado desejado.

**Nota:** É crucial notar que, ao criar uma função recursiva, é essencial ter uma condição de paragem clara e alcançável. A ausência de tal condição pode levar a um uso excessivo de memória ou de recursos de processamento, eventualmente causando falhas no programa.

No entanto, quando usada de maneira apropriada, a recursividade pode ser uma abordagem altamente eficiente e levar a soluções elegantes.

```python
#n! = n * n-1 * n-2 ... *1 for example 4! = 4 * 3 * 2 * 1
def factorial(n):
    if(n>0):
        result = n * factorial(n-1)
    else:
        result = 1
    return result

print("Factorial é",factorial(3))
```

## Funções Lambda

Funções *lambda* são **funções pequenas** e **anónimas** em Python que podem ter um qualquer número de argumentos, mas apenas uma expressão.

Estas funções são definidas usando a palavra-chave lambda e são úteis quando se precisa de uma função simples para realizar uma tarefa específica.

A sintaxe básica de uma função lambda é a seguinte:

**lambda** argumentos: expressão

```python
add = lambda x, y: x + y
print(add(5, 3))  # Saída: 8
```

```python
pessoas = [{'nome': 'Alice', 'idade': 30}, {'nome': 'Bob', 'idade': 25}, {'nome': 'Charlie', 'idade': 35}]
pessoas.sort(key=lambda x: x['idade'])
print(pessoas)
# Saída: [{'nome': 'Bob', 'idade': 25}, {'nome': 'Alice', 'idade': 30}, {'nome': 'Charlie', 'idade': 35}]
```

```python
números = [1, 2, 3, 4, 5]
quadrados = list(map(lambda x: x**2, números))
print(quadrados)
# Saída: [1, 4, 9, 16, 25]
```

```python
números = [1, 2, 3, 4, 5, 6, 7, 8, 9]
pares = list(filter(lambda x: x % 2 == 0, números))
print(pares)
# Saída: [2, 4, 6, 8]
```

# Introdução às ***classes***

Como sabemos, Python é uma linguagem orientada a objetos.

Para dominar a programação em Python, não podemos só conhecer as suas estruturas de dados e estruturas de controlo. Devemos também aprender a criar os nossos próprios objetos.

É possível criar uma classe de forma aberta como um padrão ou modelo, para a criação de objetos específicos.

Essa estrutura contém as informações, atributos e funções que serão herdadas pelos objetos criados com base nesse modelo.

## Conceitos

**Objeto**: Quando falamos em objetos, estamos a referir-nos a entidades que possuem um estado e um comportamento. Essas entidades têm a capacidade de armazenar informações e executar tarefas.

**Atributos e métodos**: Os objetos desempenham a função de armazenar dados e realizar ações. Atributos são as propriedades ou características que configuramos para o objeto, enquanto os métodos são as tarefas que eles são capazes de realizar.

## Diferenças entre métodos e funções

**Métodos:**

- Métodos são funções que estão associadas a objetos em Python. São chamados em relação a um objeto específico e operam sobre os dados contidos nesse objeto;
- Métodos são invocados usando a notação de ponto, como **objeto.metodo()**.


**Funções:**

- Funções são blocos independentes de código que podem ser chamados em qualquer lugar num programa Python;
- Funções não estão ligadas a objetos específicos e podem ser invocadas com base em seu nome;
- São definidas usando a palavra-chave **def** e são chamadas simplesmente pelo nome da função, como **funcao()**.
- Funções podem receber argumentos como entrada e retornar valores como saída.

## Criar uma classe

Para criar uma classe em Python, usamos o comando **class**, seguido do **nome** que desejamos dar à classe, e com o qual a invocaremoss; juntamente com parêntesiss, dentro dos quais poderíamos inserir o nome de outra classe para **herdar** seus atributos.

```python
class Cao():

    #Construtor
    def __init__(self, colour, breed, has_collar):
        self.colour = colour
        self.breed = breed
        self.has_collar = has_collar

    def latir(self):
        print('Woof, a minha cor é {}'.format(self.colour))
    def correr(self):
        print('Quando corro, a minha coleira cai.')
        self.has_collar = False
    def caminhar(self, x=1):
        print('. . . . ')
        print(x + 1)
    def farejar(self):
        if self.has_collar == True:
            print('farejou')
        else:
            print('não farejou')
```

Aqui temos uma Classe chamada de **Cao**, que tem como métodos: latir, correr, caminhar e farejar.

### Herança simples e múltipla

Uma das vantagens do uso de classes é a capacidade de um objeto de uma classe poder herdar os atributos e métodos de uma outra classe.

A herança simples ocorre quando uma classe ***child*** herda os atributos e métodos de uma classe monoparental.

```python
class Student(object):
    def __init__(self, age, name):
        self.age = age
        self.name = name
```

```python
class Data_student(Student):

    def introduce(self):
        print ("I'm {}, {} years old and studying Data Science".format(self.name, self.age))
```

```python
ada = Data_student(100, 'Ada L')
ada.introduce()
```

Casos de herança múltipla em Python ocorrem quando uma *child* ou classe de *child* herda atributos e métodos de mais de um parente ou classe de parentes.

```python
class Student(object):
    def __init__(self, age, name):
        self.age = age
        self.name = name

class School(object):
    def introduceSchool (self):
        print("Studying in Nuclio Digital School")

class Datascience (Student, School):
    def introduce(self):
        print ("I'm {}, {} years old and studying Data Science".format(self.name, self.age))

ada = Datascience(100, 'Ada L')
ada.introduce()
ada.introduceSchool()
```

## criando função para transformação e criação das tabelas

```python
def ret_sqlite_type(datatype:str)->str:
    if 'varchar' in datatype:
        return 'TEXT'
    elif 'text' in datatype:
        return 'TEXT'
    elif 'date' in datatype:
        return 'TEXT'
    elif 'int' in datatype:
        return 'INTEGER'
    elif 'decimal' in datatype:
        return 'REAL'
    elif 'blob' in datatype:
        return 'BLOB'
    else:
        return 'NAO CLASSIFICADO'
```

```python
def ret_sqlite_key(key:str)->str:
    if 'PRI' in str(key):
        return 'PRIMARY KEY'
    else:
        
        return ''
```

```python
def ret_sqlite_null(null_feat:str)->str:
    if 'NO' in str(null_feat):
        return 'NOT NULL'
    else:        
        return ''
```

## criando coluna de transformacao dos valores do mysql para sqlite3

```python
for key in dict_all.keys():
    #print(key)
    dict_all[key]['SQLITE3_DATATYPE'] = dict_all[key]['Type'].apply(ret_sqlite_type)
    dict_all[key]['SQLITE3_PRIMARYKEY'] = dict_all[key]['Key'].apply(ret_sqlite_key)
    dict_all[key]['SQLITE3_NULL'] = dict_all[key]['Null'].apply(ret_sqlite_null)
    
```

# Testando Tabelas Criadas

```python
def ret_values(table_name:str)->None:
    try:
        cur.execute(f"""
            SELECT 
                *
            FROM {table_name}
        """)
        print(cur.fetchall())
    except Exception as e:
        print(e)
    
```

```python
def ret_schema(table_name:str)-> None:
    """cid      name          type     notnull  dflt_value  pk"""
    try:
        cur.execute(f"""
        pragma table_info({table_name});
        """)
        lst = cur.fetchall()
        for item in lst:
            print(item)
    except Exception as e:
        print(e)
```
